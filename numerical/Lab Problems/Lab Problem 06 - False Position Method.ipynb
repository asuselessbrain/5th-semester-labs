## False Position Method

f(x)= (668.06/x)*(1− e^(−0.146843*x)) − 40

```python
import matplotlib.pyplot as plt
import numpy as np

# Defining Function
def f(x):
    fX = (0.6 * x**2) + (2.4 * x) +5.5
    return fX

# Implementing False Position Method
def falsePosition(x0,x1,e):
    step = 1
    print('\nFALSE POSITION METHOD IMPLEMENTATION: \n')
    condition = True
    x2_values = []
    f_x2_values = []
    while condition:
        x2 = x0 - (x1-x0) * f(x0)/( f(x1) - f(x0) )
        print('Iteration-%d, x2 = %0.6f and f(x2) = %0.6f' % (step, x2, f(x2)))
        x2_values.append(x2)
        f_x2_values.append(f(x2))

        if f(x0) * f(x2) < 0:
            x1 = x2
        else:
            x0 = x2

        step = step + 1
        condition = abs(f(x2)) > e

    print('\nRequired root is: %0.8f' % x2)
    return x2_values, f_x2_values

# Input Section
x0 = input('First Guess: ')
x1 = input('Second Guess: ')
e = input('Tolerable Error: ')

# Converting input to float
x0 = float(x0)
x1 = float(x1)
e = float(e)

# Checking Correctness of initial guess values and false positioning
if f(x0) * f(x1) > 0.0:
    print('Given guess values do not bracket the root.')
    print('Try Again with different guess values.')
else:
    x2_values, f_x2_values = falsePosition(x0,x1,e)

# Plotting the results
plt.figure(figsize=(15, 15))
plt.plot(x2_values, f_x2_values, marker='o')
plt.title('False Position Method')
plt.xlabel('x2 values')
plt.ylabel('f(x2) values')
plt.grid(True)
plt.show()
```
